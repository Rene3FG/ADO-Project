```python
from db_client import engine
from sqlalchemy import text

print("Probando conexión a PostgreSQL...")

try:
    with engine.connect() as conn:
        result = conn.execute(text("SELECT version();")).fetchone()
        print("\n¡Conexión exitosa a la Base de Datos local!")
        print(f"Info del servidor: {result[0]}")
except Exception as e:
    print(f"\nError al conectar a la BD. Revisa el usuario y contraseña en tu .env: {e}")
```

```python
import pandas as pd
from sheets_client import SheetsClient

print("--- 🚿 Investigando el área de Lavado Exterior ---")
sheets = SheetsClient()

try:
    datos_lavado = sheets.read_range("LAVADO EXTERIOR", "A2:F15")
    
    columnas = ["Serie_Autobus", "Hora_Entrada", "Completado", "Hora_Salida", "Hora_Entrada_Dec", "Espacios_Disp"]
    df_lavado = pd.DataFrame(datos_lavado, columns=columnas)
    
    print("Unidades actualmente registradas en Lavado:")
    display(df_lavado)
except Exception as e:
    print(f"Error al leer Lavado Exterior: {e}")
```

```python
print("--- 📅 Consultando el Itinerario de Corridas ---")
try:
    datos_plan = sheets.read_range("PLANEACION", "A2:K100")
    df_plan = pd.DataFrame(datos_plan)
    df_salidas = df_plan.iloc[:, [0, 1, 2]]
    df_salidas.columns = ["Hora_Programada", "Numero_Unidad", "Tipo_Servicio"]
    
    print("Próximos autobuses programados para salir:")
    display(df_salidas)
except Exception as e:
    print(f"Error al leer Planeación: {e}")
```
